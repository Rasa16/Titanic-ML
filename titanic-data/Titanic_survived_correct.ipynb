```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as pl
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("F:\\STUDY MATERIAL\\MACHINE LEARNING\\DATASETS\\titanic-data\\train_clean.csv")
df.head()
```

```python
test = pd.read_csv("F:\\STUDY MATERIAL\\MACHINE LEARNING\\DATASETS\\titanic-data\\test_clean.csv")
test.head()
```

```python
df.isna().sum()
```

```python
df.drop('Cabin',axis ='columns', inplace = True)
test.drop('Cabin',axis ='columns', inplace = True)
```

```python
df.drop('Name',axis ='columns', inplace = True)
df.drop('Ticket',axis ='columns', inplace = True)

test.drop('Name',axis ='columns', inplace = True)
test.drop('Ticket',axis ='columns', inplace = True)
```

```python
test.head()
```

```python
df.shape
```

```python
df.replace({'male':1, 'female':0},inplace = True)
test.replace({'male':1, 'female':0},inplace = True)
```

```python
df.head()
```

```python
test.head()
```

```python
#one hot encoding
dummies = pd.get_dummies(df['Title'])
df = pd.concat([df,dummies],axis='columns')
df.drop('Title',axis ='columns', inplace = True)
```

```python
#one hot encoding
dummies = pd.get_dummies(test['Title'])
test = pd.concat([test,dummies],axis='columns')
test.drop('Title',axis ='columns', inplace = True)
```

```python
#one hot encoding
dummies = pd.get_dummies(df['Embarked'])
df = pd.concat([df,dummies],axis='columns')
df.drop('Embarked',axis ='columns', inplace = True)

dummies1 = pd.get_dummies(test['Embarked'])
test = pd.concat([test,dummies1],axis='columns')
test.drop('Embarked',axis ='columns', inplace = True)
```

```python
x = df
x.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing
scaler = MinMaxScaler()

#train
scaler.fit(x)
x = scaler.transform(x)
```

```python
x = pd.DataFrame(x)
x.head()
```

```python
scaler.fit(test)
test = scaler.transform(test)
```

```python
test = pd.DataFrame(test)
test.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x,y)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_train,model.predict(x_train)))
print(classification_report(y_train,model.predict(x_train)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,model.predict(x_test)))
print(classification_report(y_test,model.predict(x_test)))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,model_nn.predict(x_test)))
print(classification_report(y_test,model_nn.predict(x_test)))
```
